# Data Cleaning

```python
# Checking_Missing_Values
df.isnull().sum()[df.isnull().sum()>0]
```

```python
# Fill Missing Value
# We are filling Median in missing value, if we fill mean it will affect the data
df['Review Rating']=df['Review Rating'].fillna(df['Review Rating'].median())
```

```python
df.isnull().sum()
```

```python
# Convert columns into snake case format 
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# feature engineering
# create a new column
```

```python
# Feature Engineering
# Create a new columns to divide age categories
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head()
```

```python
# Another feature
# Convert textual to Number 
df['frequency_of_purchases'].unique
```

```python
# create column name as frequency_of_purchases
# we create dict that match freq of number of days
frequency_mapping = {'Fortnightly':14,'Weekly':7,'Monthly':30,'Quarterly':90,
                     'Bi-Weekly':14, 'Annually':365, 'Every 3 Months':90 }

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# we will drop one column name as promo_code_used
# discount_applied and promo_code_used contain same values in all rows
df = df.drop(['promo_code_used'],axis=1)
```
